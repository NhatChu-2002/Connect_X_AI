```python
import matplotlib.pyplot as plt

batch_sizes = [32, 64, 128, 256]
win_rates_dqn = [14, 29, 7.5, 2.5]
win_rates_ppo = [5, 6, 4, 3.5]

plt.figure(figsize=(8, 5))
plt.plot(batch_sizes, win_rates_dqn, marker='o', linestyle='-', linewidth=2, label='Double DQN')
plt.plot(batch_sizes, win_rates_ppo, marker='s', linestyle='-', linewidth=2, label='PPO')
plt.title('Win Rate vs Batch Size: Double DQN vs PPO')
plt.xlabel('Batch Size')
plt.ylabel('Win Rate (%)')
plt.xticks(batch_sizes)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('win_rate_comparison_plot.png')

plt.show()
```

```python

epochs = [2, 4, 6, 8, 10]
win_rates_ppo_epochs = [4, 6, 11, 12.5, 9.5]

plt.figure(figsize=(8, 5))
plt.plot(epochs, win_rates_ppo_epochs, marker='d', linestyle='-', linewidth=2, color='teal', label='PPO')
plt.title('PPO: Win Rate vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Win Rate (%)')
plt.xticks(epochs)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('ppo_win_rate_epochs_plot.png')

plt.show()
```

```python
episodes = [10000, 30000, 50000, 100000, 1000000]
win_rates_dql = [10.5, 21, 31.5, 29, 31.5]

# Bigger figure size (width=12, height=7)
plt.figure(figsize=(12, 7))
plt.plot(episodes, win_rates_dql, marker='o', linestyle='-', linewidth=2, color='orange', label='Double DQN')
plt.title('Double DQN: Win Rate vs Training Episodes', fontsize=18)
plt.xlabel('Training Episodes', fontsize=14)
plt.ylabel('Win Rate (%)', fontsize=14)
plt.xticks(episodes, rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.tight_layout()

# Save and show
plt.savefig('double_dqn_win_rate_episodes_plot_large.png')
plt.show()
```

```python

timesteps = [500_000, 1_000_000, 2_000_000, 5_000_000, 10_000_000]
win_rates_ppo = [4, 6, 14.5, 32, 43]

# Create plot with larger figure size
plt.figure(figsize=(12, 7))
plt.plot(timesteps, win_rates_ppo, marker='s', linestyle='-', linewidth=2, color='teal', label='PPO')
plt.title('PPO: Win Rate vs Training Timesteps', fontsize=18)
plt.xlabel('Training Timesteps', fontsize=14)
plt.ylabel('Win Rate (%)', fontsize=14)
plt.xticks(timesteps, [f'{int(t/1e6)}M' if t >= 1e6 else f'{int(t/1e3)}K' for t in timesteps], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend(fontsize=12)
plt.tight_layout()

# Save the figure
plt.savefig('ppo_win_rate_timesteps_plot.png')

# Show the plot
plt.show()
```
